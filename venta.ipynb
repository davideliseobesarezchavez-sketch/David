```
from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
import numpy as np
from tensorflow import keras
from sklearn.linear_model import LinearRegression
```

```
# Cargar el dataset Fashion MNIST
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Definir las etiquetas de las categorías
labels = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
          "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

# Explorar el dataset dibujando las primeras 20 imágenes
plt.figure(figsize=(14, 8))
for i in range(20):
    plt.subplot(4, 5, i+1)
    plt.imshow(X_train[i], cmap="binary")
    plt.title(labels[y_train[i]])
    plt.axis("off")
plt.show()
```

```
# Normalizar los datos escalando los píxeles al rango [0, 1]
X_train, X_test = X_train / 255.0, X_test / 255.0
print("Datos normalizados correctamente.")
```

```
# 1. Definir la arquitectura
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

# 2. Compilar el modelo
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 3. Entrenar el modelo (10 épocas)
print("Iniciando entrenamiento...")
historial = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

```
# Evaluar el modelo en el conjunto de prueba
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(f'\n==================================================')
print(f'🎯 Precisión final en el conjunto de prueba: {test_acc * 100:.2f}%')
print(f'==================================================')
```

```
# Paso 1: Fijar una semilla matemática para que los datos sean reproducibles
np.random.seed(42)

# Paso 2: Generar 50 datos aleatorios para la variable independiente (X)
X = 2 * np.random.rand(50, 1)

# Paso 3: Generar la variable dependiente (y) basada en la fórmula: y = 4 + 3x + ruido
# Esto simula un escenario real (ej. relación entre Temperatura y Ventas)
y = 4 + 3 * X + np.random.randn(50, 1)

# Paso 4: Crear e instanciar el algoritmo de Regresión Lineal
modelo_lineal = LinearRegression()

# Paso 5: Entrenar el modelo para ajustar la línea recta a los datos
modelo_lineal.fit(X, y)
```

```
# Paso 6: Imprimir en pantalla los parámetros calculados por la Inteligencia Artificial
print("==================================================")
print("📊 RESULTADOS DEL MODELO DE REGRESIÓN")
print("==================================================")
print(f"Intercepto calculado (b)  - Punto donde cruza el eje Y: {float(modelo_lineal.intercept_):.4f}")
print(f"Pendiente calculada (m)  - Inclinación de la recta:     {float(modelo_lineal.coef_[0]):.4f}")
print("==================================================\n")

# Paso 7: Definir los puntos extremos en el eje X para realizar las predicciones de la línea
X_nuevos = np.array([[0], [2]])
y_predicciones = modelo_lineal.predict(X_nuevos)

# Paso 8: Graficar los resultados en Google Colab
plt.figure(figsize=(9, 6))
```

```
# Dibujar los puntos azules (muestras de datos reales con ruido)
plt.scatter(X, y, color='#1f77b4', alpha=0.8, edgecolors='k', s=50, label='Datos reales (Muestras)')

# Dibujar la línea roja (la predicción perfecta que calculó el modelo)
plt.plot(X_nuevos, y_predicciones, color='#d62728', linewidth=2.5, label='Línea de regresión predictiva')

# Configuraciones estéticas de la gráfica
plt.xlabel('Variable Independiente (X)', fontsize=12)
plt.ylabel('Variable Dependiente (y)', fontsize=12)
plt.title('Modelo de Regresión Lineal Simple en Google Colab', fontsize=14, fontweight='bold')
plt.legend(loc='upper left', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar el resultado visual en Colab
plt.show()
```
